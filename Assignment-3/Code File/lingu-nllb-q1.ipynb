```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")

```

```python
article = "UN Chief says there is no military solution in Syria"
inputs = tokenizer(article, return_tensors="pt")

translated_tokens = model.generate(
    **inputs, forced_bos_token_id=tokenizer.lang_code_to_id["guj_Gujr"], max_length=30
)
tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
```

```python
file_path = "/kaggle/working/hi_1000_ground_truth.txt"
#load english dataset
with open(file_path, "r", encoding="utf-8") as file:
    hi_1000_text = file.readlines()

print("Number of lines in the dataset:", len(hi_1000_text))

# print(hi_text)
```

```python
model=model.to('cuda')
```

English to Mother Tounge (Gujarati) Translation

```python
translations = [] # list storing translation

for line in en_1000_text:
  inputs = tokenizer(line, return_tensors="pt") # tokenize input
  inputs = {key: value.to(model.device) for key, value in inputs.items()}

  translated_tokens = model.generate(
    **inputs, forced_bos_token_id=tokenizer.lang_code_to_id["guj_Gujr"], max_length=200
  )
  translated_text=tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
  translations.append(translated_text)
```

```python
translations[0]
```

```python

Eng_guj_tran = 'eng_guj_tran_NLLB.txt'
# Open the output file in write mode
with open(Eng_guj_tran, "w", encoding="utf-8") as file:
    for translation in translations:
        # Write each translation to the file
        file.write(translation + "\n")

print("Translations file :",Eng_guj_tran)
```

Gujarati to English Translation

```python
eng_translations = [] # list storing translation from gujarati to english
#eng_latn for english translation
for line in gu_1000_text:
  inputs = tokenizer(line, return_tensors="pt") # tokenize input
  inputs = {key: value.to(model.device) for key, value in inputs.items()}

  translated_tokens = model.generate(
    **inputs, forced_bos_token_id=tokenizer.lang_code_to_id["eng_Latn"], max_length=100
  )
  translated_text=tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
  eng_translations.append(translated_text)
```
